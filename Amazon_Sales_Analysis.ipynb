# Sales Analysis

```python
sales_by_state = df.groupby("State")["TotalAmount"].sum()
sales_by_state
```

```python
sales_by_state.sort_values(ascending=False)
```

## Sales by Category

```python
sales_by_category = df.groupby("Category")["TotalAmount"].sum()
sales_by_category
```

```python
sales_by_category.sort_values(ascending=False)
```

## Top 10 Products by Sales

```python
top_products = df.groupby("ProductName")["TotalAmount"].sum()
top_products
```

```python
top_products.sort_values(ascending=False).head(10)
```

## Top 5 Customers by Sales

```python
top_customers = df.groupby("CustomerName")["TotalAmount"].sum()
top_customers
```

```python
top_customers.sort_values(ascending=False).head(5)
```

## Payment Mode Analysis

```python
payment_mode = df["PaymentMethod"].value_counts()
payment_mode
```

## Monthly Sales Trend

```python
df["OrderDate"] = pd.to_datetime(df["OrderDate"])
```

```python
df["Month"] = df["OrderDate"].dt.strftime("%B")
```

```python
monthly_sales = df.groupby("Month")["TotalAmount"].sum()
monthly_sales
```

## Sales by State Chart

```python
import matplotlib.pyplot as plt
sales_by_state.sort_values(ascending=False).plot(kind="bar", figsize=(10,5))
plt.title("Sales by State")
plt.xlabel("State")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

## Sales by Category (Pie Chart)

```python
sales_by_category.plot(
    kind="pie",
    autopct="%1.1f%%",
    figsize=(7,7)
)
plt.title("Sales by Category")
plt.ylabel("")
plt.show()
```

## Monthly Sales Trend (Line Chart)

```python
monthly_sales.plot(
    kind="line",
    marker="o",
    figsize=(10,5)
)

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)

plt.show()
```

## Top 10 Products by Sales (Bar Chart)

```python
top_products.sort_values(ascending=False).head(10).plot(
    kind="barh",
    figsize=(10,6)
)

plt.title("Top 10 Products by Sales")
plt.xlabel("Total Sales")
plt.ylabel("Product Name")

plt.show()
```
